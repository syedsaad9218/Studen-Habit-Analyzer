```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('student_habits.csv')

df.head()
```

```python
sns.barplot(data=df, x='study_hours', y='sleep_hours')
plt.title('Study Hours vs Sleep Hours')
plt.xlabel('Study Hours')
plt.ylabel('Sleep Hours')
plt.show()
```

```python
sns.barplot(data=df, x='phone_usage', y='exercise_time',)
plt.title('Phone Usage vs Exercise Time')
plt.xlabel('Phone Usage (hours)')
plt.ylabel('Exercise Time (hours)')
plt.show()
```

```python
sns.barplot(data=df, x='performance', y='attendance')
plt.title('Performance vs Attendance')
plt.xlabel('Performance')
plt.ylabel('Attendance')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
fig = px.scatter_3d(df, x='study_hours', y='sleep_hours', z='performance', color='attendance', size='exercise_time', title='3D Scatter Plot of Student Habits')
fig.show()
```

```python
sns.barplot(data=df, x='performance', y='study_hours')
plt.title('Performance vs Study Hours')
plt.xlabel('Performance')
plt.ylabel('Study Hours')
plt.show()
```
